```python
%%writefile workout_planner.py

def user_input(message):
    while True:
        try:
            value = float(input(message))
            if value <= 0:
                raise ValueError
            return value
        except ValueError:
            print("Invalid input! Please enter a positive number.\n")

def calculate_bmi(weight_in_kg, height_in_cm):
    height_in_meters = height_in_cm / 100
    return weight_in_kg / (height_in_meters * height_in_meters)

def suggest_plan(bmi):
    if bmi < 18.5:
        return "Workout: 20 minutes/day", "Diet: High-calorie balanced diet", "Location: Home workouts"
    elif bmi < 25:
        return "Workout: 30 minutes/day", "Diet: Balanced diet", "Location: Home or Gym"
    else:
        return "Workout: 45 minutes/day", "Diet: Low-fat, healthy diet", "Location: Gym workouts"
```

```python
from workout_planner import user_input, calculate_bmi, suggest_plan

weight_in_kg = user_input("Enter your weight in kg: ")
height_in_cm = user_input("Enter your height in cm: ")
bmi = calculate_bmi(weight_in_kg, height_in_cm)
workout, diet = suggest_plan(bmi)
print("Your BMI is:", bmi)
print(workout)
print(diet)
```
